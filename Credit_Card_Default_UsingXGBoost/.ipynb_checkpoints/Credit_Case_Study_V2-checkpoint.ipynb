# Converting columns to bool

```python
df['owns_car'] = df['owns_car'].apply(lambda x: 1 if x== 'Y' else 0)
df['owns_house']= df['owns_house'].apply(lambda x: 1 if x== 'Y' else 0)
df['gender'] = df['gender'].apply(lambda x: 1 if x== 'M' else 0)
```

```python
df.head(2)
```

```python
for i in df.columns:
    if i in ['occupation_type', 'no_of_children', 'total_family_members']:
        myd= pd.get_dummies(df[i], prefix= i, prefix_sep= '_', drop_first= True)
        df= pd.concat([df, myd], axis= 1)
        df.drop(labels= i, axis= 1, inplace= True)
        
```

# RFE

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
logreg= LogisticRegression()
```

```python
rfe= RFE(logreg, 15)
rfe= rfe.fit(X_train_ml, y_train)
```

```python
rfe_table= pd.DataFrame()
rfe_table['Feature']= X_train_ml.columns
rfe_table['Retain']= rfe.support_
rfe_table['Rank']= rfe.ranking_
rfe_table
```

```python
rfe_table = rfe_table[rfe_table['Retain']== True]
```

```python
rfe_table
```

```python
X_train_ml= X_train_ml[rfe_table['Feature']]
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Define GLM Table code

def glm_table(X, y):
    logmodel= sm.GLM(y, sm.add_constant(X), family= sm.families.Binomial())
    result= logmodel.fit()
    print(result.summary())

# Define vif table code

def vif_table(X_train):
    vif_table= pd.DataFrame()
    vif_table['Feature']= X_train.columns
#     vif_table['VIF Value']= [variance_inflation_factor(X_train[col].values, i) for i in range(X_train[col].shape[1])]
    vif_table['VIF Value']= [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
    vif_table= vif_table.sort_values(by= 'VIF Value', ascending= False)
    print(vif_table)
```
